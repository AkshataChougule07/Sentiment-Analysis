```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}

for url_id,url in zip(df['URL_ID'],df['URL']):
    #print(url)
    source=requests.get(url,headers=headers)
    
    soup=BeautifulSoup(source.content, 'html.parser')
    #print(url)
    #print(soup.title.text)
    #print("\n")
    
    ff = open(str(url_id)+'.txt', "w",encoding='utf-8')
    ff.write(soup.title.text)
    ff.close()

    ff=open(str(url_id)+'.txt',"a",encoding='utf-8')
    for doc in soup.find_all("div",attrs={'class':'td-post-content'}):
        ff.write(doc.get_text())   
    ff.close()
```

Calculating Positive and Negative Score

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import io
import string

#df_pos_neg=pd.DataFrame(columns=['POSITIVE SCORE','NEGATIVE SCORE'])
result_pos=[]
result_neg=[]

for w in df['URL_ID']:
      
    input_str=""
    with open(str(w)+".txt","r",encoding="utf8") as f:
        for line in f.read():
            input_str=input_str+line.lower()      #reading file as a lower case letters
            cleaned_text = input_str.translate(str.maketrans('', '', string.punctuation))       #removing punctuations


    f.close()
    
    stop_words = set(stopwords.words('english'))
    token=word_tokenize(cleaned_text)
    
   
    final_words = []              #list contains cleaned data
    for word in token:
        if word not in stopwords.words('english'):
            final_words.append(word)
    #print(w," ",final_words)
    pos=0
    ne=0
    
    for word in final_words:
        if word in pos_list:
            pos+=1
        if word in neg_list:
            ne-=1
    pos_score=pos
    neg_score=ne*-1
    result_pos.append(pos_score)
    result_neg.append(neg_score)


df['POSITIVE SCORE']=result_pos
df['NEGATIVE SCORE']=result_neg 

       
```

Percentage of Complex Words

```python
len_words=[]
complex_per=[]

for v in df_black['URL_ID']:   
    input_str=""
    with open(str(v)+".txt","r",encoding="utf8") as f:
        for line in f.read():
            input_str=input_str+line.lower()      #reading file as a lower case letters
            cleaned_text = input_str.translate(str.maketrans('', '', string.punctuation))


    f.close()
    
    stop_words = set(stopwords.words('english'))
    token=word_tokenize(cleaned_text)
   
    final_words = []

    for word in token:
        if word not in stopwords.words('english'):
            final_words.append(word)
    len_words.append(len(final_words))
    
    complexWord = 0
    complex_word_percentage = 0
    
    for word in final_words:
        vowels=0
        if word.endswith(('es','ed')):
            pass
        else:
            for w in word:
                if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u'):
                    vowels += 1
            if(vowels > 2):
                complexWord += 1
    if len(final_words) != 0:
        complex_word_percentage = complexWord/len(final_words)
        print(v," : ",round(complex_word_percentage,2))
        complex_per.append(round(complex_word_percentage,2))

df_black['PERCENTAGE OF COMPLEX WORDS']=complex_per
    
  
```
